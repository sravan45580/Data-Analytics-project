```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('noseInst.csv')
data.head()
```

```python
data.info()
```

# Load the dataset

```python
data.describe()
```

```python
missing_values = data.isnull().sum()
print("\nMissing values:\n", missing_values)
```

# Check for missing values

```python
df.columns = [
    "States",
    "Secondary_Board_Count",
    "Senior_Secondary_Schools",
    "High_Secondary_Schools",
    "Upper_Primary_Schools",
    "Primary_Schools",
    "Total_Schools",
    "Pre_Primary_Schools"
]
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Total_Schools'], kde=True, bins=15, color='blue')
plt.title('Distribution of Total Schools', fontsize=16)
plt.xlabel('Total Schools')
plt.ylabel('Frequency')
```

```python
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = data.iloc[:, 1:].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```

```python
top_states = data.nlargest(5, 'Total_Schools')
plt.figure(figsize=(10, 6))
sns.barplot(x='Total_Schools', y='States', data=top_states, palette='viridis')
plt.title('Top 5 States with the Highest Total Schools', fontsize=16)
plt.xlabel('Total Schools')
plt.ylabel('States')
plt.show()
```

```python
data['Pre_Primary_Percentage'] = (data['Pre_Primary_Schools'] / data['Total_Schools']) * 100
plt.figure(figsize=(12, 6))
sns.barplot(
    x='Pre_Primary_Percentage', y='States', data=data.sort_values('Pre_Primary_Percentage', ascending=False),
    palette='cool'
)
plt.title('Pre-Primary Schools as Percentage of Total Schools', fontsize=16)
plt.xlabel('Percentage (%)')
plt.ylabel('States')
plt.show()
```

```python
sns.pairplot(data.iloc[:, 1:])
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=data.iloc[:, 1:], orient='h', palette='Set3')
plt.title('Boxplot for Numerical Features', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sorted_data = data.sort_values(by='Total_Schools', ascending=False)
sns.lineplot(x='States', y='Total_Schools', data=sorted_data, marker='o', color='green')
plt.xticks(rotation=90)
plt.title('Trend of Total Schools by States', fontsize=16)
plt.xlabel('States')
plt.ylabel('Total Schools')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='High_Secondary_Schools', y='States', data=data, palette='muted')
plt.title('Distribution of High/Secondary Schools by States', fontsize=16)
plt.xlabel('High/Secondary Schools')
plt.ylabel('States')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Primary_Schools', y='Total_Schools', data=data, hue='States', palette='husl')
plt.title('Primary Schools vs Total Schools', fontsize=16)
plt.xlabel('Primary Schools')
plt.ylabel('Total Schools')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
data.to_csv('cleaned_noseInst.csv', index=False)
print("Cleaned dataset saved as 'cleaned_noseInst.csv'.")
```

```python
plt.figure(figsize=(12, 8))
plt.pie(
    data['Total_Schools'], labels=data['States'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'),
)
plt.title('Distribution of Total Schools by States', fontsize=16)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.pie(
    data['Pre_Primary_Schools'], labels=data['States'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('cool'),
)
plt.title('Distribution of Pre-Primary Schools by States', fontsize=16)
plt.axis('equal')
plt.show()
```

```python
statewise_sum = data.groupby('States').sum()
print("\nState-wise summary:\n", statewise_sum)
```

```python
mean_total_schools = data['Total_Schools'].mean()
high_school_states = data[data['Total_Schools'] > mean_total_schools]
print("\nStates with above average total schools:\n", high_school_states[['States', 'Total_Schools']])
```

```python
data.to_csv('cleaned_noseInst.csv', index=False)
print("Cleaned dataset saved as 'cleaned_noseInst.csv'.")
```
